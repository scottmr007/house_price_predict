### Temporary Imputation Strategy

```python
df_temp = df.copy()

# Numerical columns: Choose mean/median based on skewness
for col in df_temp.select_dtypes(include=['int64', 'float64']).columns:
    if df_temp[col].skew() < 1:  # Low skewness Use mean
        df_temp[col] = df_temp[col].fillna(df_temp[col].mean())  
    else:  # High skewness Use median
        df_temp[col] = df_temp[col].fillna(df_temp[col].median())  

# Categorical columns: Mode or 'Unknown'
for col in df_temp.select_dtypes(include=['object']).columns:
    if df_temp[col].nunique() > 10:  # Many unique values Use 'Unknown'
        df_temp[col] = df_temp[col].fillna("Unknown")  
    else:  # Few unique values Use mode
        df_temp[col] = df_temp[col].fillna(df_temp[col].mode()[0])  

# Now proceed with EDA using df_temp
```

```python
sns.set_theme(rc={'figure.figsize':(24,16)})
sns.set_theme(style="white")
sns.boxenplot(data=df_temp, x='price', y='yr_built', orient='y')
plt.title('Year and Price')
plt.ylabel('Year Built')
plt.ticklabel_format(style='plain', axis='x')
plt.xlabel('Price')

# Remove borders around plot
for spine in plt.gca().spines.values():
    spine.set_visible(False)
```

```python
df_temp = pd.DataFrame(data=df_temp)
df_temp.drop(['date'], axis=1, inplace=True)
corr = df_temp.corr()
# Generate a mask for upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))
plt.subplots(figsize=(15,8))
sns.heatmap(corr, vmin=-1, vmax=1, mask=mask)
```
